6. (7б) Визначте функцію, яка приймає iterable (ітерабельний обʼєкт, напр. список), що складається з 7ми рядкових значень - назв погодних умов (напр. "sunny") за кожен день у тижні. Повертає функція словник, де ключами є рядки з погодними умовами, а значеннями - кількість разів, скільки дана погода була протягом тижня (напр. {'sunny': 1, ...}). Викличте цю функцію та виведіть результат її роботи на екран.

```python
def count_weather_conditions(weekly_weather):
    weather_count = {}
    for condition in weekly_weather:
        if condition in weather_count:
            weather_count[condition] += 1
        else:
            weather_count[condition] = 1
    return weather_count

weekly_weather = ["rainy", "rainy", "cloudy", "sunny", "cloudy", "rainy", "cloudy"]

result = count_weather_conditions(weekly_weather)
print(result)
```

7. (15б) Напишіть програму, яка б шукала необхідне число у впорядкованому списку цілочислених значень, де
число має вводитись з консолі.
власне алгоритм пошуку винесений в окрему функцію, яка повертає булеве значення (знайдено число - True, не знайдено - False).
в залежності від результату функції, виведіть на екран повідомлення про успішний або ж неуспішний пошук.
Зауваження: використовувати оператор in заборонено.

```python
def binary_search(sorted_list, target):
    left = 0
    right = len(sorted_list) - 1

    while left <= right:
        mid = (left + right) // 2
        mid_value = sorted_list[mid]

        if mid_value == target:
            return True
        elif mid_value < target:
            left = mid + 1
        else:
            right = mid - 1

    return False

sorted_list = [2, 5, 6, 8, 10, 11, 16]

target = int(input("Enter the num to find: "))

if binary_search(sorted_list, target):
    print("Num is found in the given list")
else:
    print("Num can't be found in the given list")
```

8. (4б) В яких випадках варто використовувати лямбда-функції?


У функціях вищих порядків: Лямбда-функції часто використовуються як аргументи в функціях вищих порядків, таких як map(), filter() або sorted(). Вони дозволяють зручно визначати короткі функції, не використовуючи додаткове ім'я.

При необхідності коротких виразів: Лямбда-функції зазвичай коротші і простіші, ніж звичайні функції, тому їх варто використовувати, коли вам потрібно визначити швидку і просту функцію без багатої інфраструктури.

У сортуванні або групуванні даних: Ви можете використовувати лямбда-функції для визначення критеріїв сортування або групування даних за певними параметрами.

У моментальних обчисленнях: Якщо вам потрібно визначити невелику функцію в місці її використання, наприклад, при використанні методу sort() або sorted() для списків, лямбда-функції дозволяють це зробити без додаткових оголошень функцій.

# III. Decorators.

11. (11б) Визначте декоратор, який би виводив у консоль повідомлення про початок роботи і назву функції (напр. my_function._ _ name _ _), а після відпрацювання функції - повідомлення про закінчення роботи і назву функції, що задекорована.

```python
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print("Start of the running function:", func.__name__)
        result = func(*args, **kwargs)
        print("End of the running func:", func.__name__)
        return result
    return wrapper

@my_decorator
def example_function(x, y): print(x + y)

example_function(3, 5)
```
